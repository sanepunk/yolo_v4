```python
import os
from glob import glob

# Define the folders
images_dir = "/content/image_chips_native"
labels_dir = "/content/labels"

image_files = glob(os.path.join(images_dir, "*.jpg"))
label_dict = {os.path.splitext(os.path.basename(l))[0]: l for l in glob(os.path.join(labels_dir, "*.txt"))}

# Pair images with corresponding label file
paired_files = []
for img in image_files:
    base = os.path.splitext(os.path.basename(img))[0]
    if base in label_dict:
        paired_files.append((img, label_dict[base]))
    else:
        print(f"Warning: No label found for {img}")


print(f"Found {len(paired_files)} paired files.")
```

```python
from sklearn.model_selection import train_test_split

# Assuming paired_files is your list of (image, label) tuples
paired_files.sort()  # Optional: to maintain a consistent order

train_pairs, test_pairs = train_test_split(paired_files, test_size=0.2, random_state=42)

print(f"Training pairs: {len(train_pairs)}")
print(f"Test pairs: {len(test_pairs)}")
```

```python
import os
import shutil

# Define output directories
output_dir = "dataset_split"
train_img_dir = os.path.join(output_dir, "train/images")
train_lbl_dir = os.path.join(output_dir, "train/labels")
test_img_dir  = os.path.join(output_dir, "test/images")
test_lbl_dir  = os.path.join(output_dir, "test/labels")

os.makedirs(train_img_dir, exist_ok=True)
os.makedirs(train_lbl_dir, exist_ok=True)
os.makedirs(test_img_dir, exist_ok=True)
os.makedirs(test_lbl_dir, exist_ok=True)

def copy_pairs(pairs, img_dest, lbl_dest):
    for img_path, lbl_path in pairs:
        shutil.copy(img_path, os.path.join(img_dest, os.path.basename(img_path)))
        shutil.copy(lbl_path, os.path.join(lbl_dest, os.path.basename(lbl_path)))

# Copy training files
copy_pairs(train_pairs, train_img_dir, train_lbl_dir)
# Copy test files
copy_pairs(test_pairs, test_img_dir, test_lbl_dir)
```

```python
def create_file_list(folder, output_file):
    image_paths = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".jpg")]
    image_paths.sort()  # Optional, for consistency
    with open(output_file, "w") as f:
        for path in image_paths:
            f.write(path + "\n")

create_file_list(train_img_dir, os.path.join(output_dir, "train.txt"))
for filename in os.listdir(train_lbl_dir):
    if filename.endswith(".txt"):
        src = os.path.join(train_lbl_dir, filename)
        dst = os.path.join(train_img_dir, filename)
        shutil.move(src, dst)
create_file_list(test_img_dir, os.path.join(output_dir, "test.txt"))
for filename in os.listdir(test_lbl_dir):
    if filename.endswith(".txt"):
        src = os.path.join(test_lbl_dir, filename)
        dst = os.path.join(test_img_dir, filename)
        shutil.move(src, dst)
```

```python
if os.path.exists("darknet"):
    %cd ..
    !rm -rf darknet
!git clone https://github.com/AlexeyAB/darknet
%cd darknet
!wget https://github.com/AlexeyAB/darknet/releases/download/darknet_yolo_v3_optimal/yolov4.conv.137
os.mkdir('/content/darknet/data/backup')
# !./darknet detector test cfg/coco.data cfg/yolov4.cfg yolov4.weights data/dog.jpg
def modify_yolov4_cfg(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Iterate through the lines and look for the section with the classes parameter
    for i, line in enumerate(lines):
      if i + 2 < len(lines) and lines[i].startswith("# Training") and lines[i+1].startswith('batch='):
        lines[i+1] = "batch=32\n"
        lines[i+2] = "subdivisions=32\n"
      if i + 7 < len(lines) and lines[i+7].startswith("classes="):
        lines[i] = "filters=24\n"
      if line.startswith("classes="):
          lines[i] = "classes=3\n"

    # Write the modified lines back to the file
    with open(file_path, 'w') as file:
        file.writelines(lines)

# Provide the path to your yolov4.cfg file
file_path = 'cfg/yolov4-custom.cfg'

# Call the function to modify the file
modify_yolov4_cfg(file_path)
```

```python
file_path = '/content/darknet/Makefile'

def modify_makefile(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Iterate through the lines and look for the section with the classes parameter
    for i, line in enumerate(lines):
        if line.startswith("GPU="):
            lines[i] = "GPU=1\n"
        if line.startswith("CUDNN="):
            lines[i] = "CUDNN=1\n"
        if line.startswith("CUDNN_HALF="):
            lines[i] = "CUDNN_HALF=1\n"
        if line.startswith("OPENCV="):
            lines[i] = "OPENCV=1\n"

    # Write the modified lines back to the file
    with open(file_path, 'w') as file:
        file.writelines(lines)


# Call the function to modify the file
modify_makefile(file_path)
```
